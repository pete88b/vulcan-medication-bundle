this is a set of synthetically created EHR data on a FHIR server (via MITRE’s Synthea).  Now, we know the data is synthetically created by elaborate algorithms – no one is going to do research on it.  But it is likely to be very nicely populated FHIR data the way EHR’s probably *should* be but aren’t.  And of course it’s very testable data probably in pretty good shape, which is what we need.  

```python
from vulcan_medication_bundle.core import *
from vulcan_medication_bundle.per_patient import *
```

```python
api_base = 'https://demo1.aha.accenture.com/fhir/r4'
```

```python
bundle = get_bundle_as_raw_json(api_base, 'ResearchSubject')
references = extract_references(bundle, ['individual'])
references
```

```python
page_count,page_limit = 0,10
bundle = get_bundle_as_raw_json(api_base, 'MedicationRequest')
references = extract_references(bundle, ['subject'])
while bundle is not None:
    page_count+=1
    if page_count>page_limit:
        print('Stopping early')
        break
    bundle = get_next_bundle_as_raw_json(bundle)
    references.extend(extract_references(bundle, ['subject']))
set(references)
```

```python
page_count,page_limit = 0,10
bundle = get_bundle_as_raw_json(api_base, 'MedicationAdministration')
references = []
while bundle is not None:
    references.extend(extract_references(bundle, ['subject']))
    page_count+=1
    if page_count>page_limit:
        print('Stopping early')
        break
    bundle = get_next_bundle_as_raw_json(bundle)
set(references)
```

```python
for p in references:
    p=p.split('/')[1]
    print(f"""        <li>
          <a onclick="useExampleValues('{api_base}', '{p}');">{p}</a>
        </li>""")
```
